```python
import pandas as pd

filename = '../Current/Data/taxi.csv'
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
df.columns
```

```python
# if I want to retrieve selected columns, I can do that with []
df['passenger_count']
```

```python
# what if I want two columns (or more)?
# then I can put a list of columns (list of strings) inside of the square brackets
# yes, that means nested square brackets

df[['passenger_count', 'total_amount', 'trip_distance']]
```

```python
# I can also use the "filter" method to indicate which of the 
# columns I want to retrieve

df.filter(['passenger_count', 'total_amount', 'trip_distance'])
```

```python
# what if I want all of the columns that have "amount" in their names?

df.filter([one_column
  for one_column in df.columns
  if 'amount' in one_column])
```

```python
# I can use "filter" and the "like" keyword argument to it!

df.filter(like='amount')
```

```python
df.filter(like='dropoff')
```

```python
df.filter(like='pickup')
```

```python
# what if I want all of the columns for *both* pickup and dropoff?
# we can use a regular expression!

df.filter(regex='pickup|dropoff')    # either "pickup" or "dropoff"
```

```python
df.filter(regex='.*tude$')
```

```python
df.head()
```

```python
# set the index to be the pickup datetime -- note this is a string!
df = df.set_index('tpep_pickup_datetime')
```

```python
df.head()
```

```python
# I want to find the *rows* that match a particular string

df.filter(like='2015-06-02')  # not what we want!
```

```python
df.filter(like='2015-06-02', axis='rows')
```

```python
df.filter(like=' 11:', axis='rows')
```

```python
df.filter(regex=r' (\d\d):\1:', axis='rows')   # space, two digits, a colon, and then THE SAME two digits
```
