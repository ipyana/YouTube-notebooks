```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
np.random.seed(0)

df = DataFrame(np.random.randint(0, 10, [4, 5]),
              index=list('wxyz'),
              columns=list('abcde'))
df
```

```python
df['c'].replace(3, 99)
```

```python
df.replace(3, 99)
```

```python
# 6 --> 96, 7 --> 97, 8 --> 98

df.replace([6, 7, 8], [96, 97, 98])
```

```python
df.replace({6:96, 7:97, 8:98})
```

```python
# turn 6 -> 96 in column b
# turn 7 -> 97 in column c
# turn 8 -> 98 in column d

df.replace({'b':{6:96}, 'c':{7:97}, 'd':{8:98}})
```

```python
# keep the values that I've set in df.replace -- assign the result back to the original variable

df = df.replace({'b':{6:96}, 'c':{7:97}, 'd':{8:98}})
```
