```python
# let's make our index have year, season, and sport

df = df.reset_index()
df = df.set_index(['Year', 'Season', 'Sport'])
df = df.sort_index()
df.head()
```

```python
# I can retrieve by year
df.loc[1992]
```

```python
# I can retrieve by year + season

df.loc[(1992, 'Summer')]
```

```python
# retrieve by three parts of the multi-index
df.loc[(1992, 'Summer', 'Archery')]
```

```python
# what if I want to find all of the people who participated in Archery across all years and seasons?
# basically: any year, any season, but just Archery

df.xs('Archery', level='Sport')
```

```python
# Let's find the average height of people in Archery over all years of the Olympics

df.xs('Archery', level='Sport')['Height'].mean()
```

```python
df.xs('Archery', level='Sport')[['Height', 'Weight']].mean()
```

```python
# what if I want to specify more than one thing with xs?
# for example: I want 1992, all seasons, archery

df.xs((1992, 'Archery'), level=['Year', 'Sport'])[['Height', 'Weight']].mean()
```
