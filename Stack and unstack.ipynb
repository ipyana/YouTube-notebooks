```python
import pandas as pd
```

```python
df = pd.read_csv('taxi.csv')
```

```python
# I'm going to run a very complex groupby

g = df.groupby(['VendorID', 'passenger_count'])[['total_amount', 'trip_distance']].agg(['mean', 'std'])
g
```

```python
# # what if I want to move the 2nd part of the column index to the rows?
# that is -- I want to to move mean+std to the multi-index on the rows, 
# making the data frame longer but narrower

# this is known as "stack"
g.stack(level=1)
```

```python
g.stack(level=0) # this means: take the outer layer of the multi-index columns, and move to the rows
```

```python
# stack is what you want to do when you have:
# a multi-index in your column
# you want to move part of that multi-index to your rows

# but what if you want to do the opposite?
```

```python
g
```

```python
# what if I want to move VendorID to the columns?
# that'll make my data frame wider, and the number of rows lower

# this is unstack!

g.unstack(level=0)
```

```python
# could I do this with passenger_count? Yes!

g.unstack(level=1)
```

```python
g
```

```python
g = g.stack(level=1)
g
```

```python
# what if I want to move more than one part of the rows' multi-index to the columns?
g.unstack(level=[0, 2])
```
