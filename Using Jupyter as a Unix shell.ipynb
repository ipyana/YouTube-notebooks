```python
all_files = !ls 
```

```python
type(all_files)
```

```python
# .l -- returns a list of strings
all_files.l
```

```python
# .n -- returns a single string, with all elements joined with \n
print(all_files.n)
```

```python
# .s -- returns a single string, with all elements joined by a space
print(all_files.s)
```

```python
# .p -- returns a list of Pathlib objects

all_files.p
```

```python
all_files = !ls -l
```

```python
all_files
```

```python
all_files.fields(0, 2, -1)
```

```python
all_files.fields(4, -1)
```

```python
# I'd like to sort my files by size, which is field 4
all_files.sort(4)
```

```python
# we need to sort column 4 (the size) as a bunch of numbers
all_files.sort(4, nums=True)
```

```python
# regular expressions -- regexps -- https://RegexpCrashCourse.com
all_files.grep('.*conf$')  # all files that end with conf
```

```python
# what if I want all of the files that end with something *OTHER* than conf?

all_files.grep('.*conf$', prune=True) 
```
